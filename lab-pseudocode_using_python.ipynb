### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

START

INPUT a positive integer and store it in N.

INITIALIZE a variable total_sum to 0.

FOR each number i from 1 up to and including N:

ADD i to total_sum.

PRINT the total_sum.

END

<!-- Add your pseudocode here -->

```python
n = int(input("Enter a positive integer (N): "))

total_sum = 0

for i in range(1, n + 1):
    total_sum += i

print(f"The sum of the first {n} natural numbers is: {total_sum}")
```

### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

START

INPUT a string and store it in text.

CONVERT text to lowercase to make it easier to compare.

DEFINE a list of vowels: a, e, i, o, u.

INITIALIZE a counter variable vowel_count to 0.

FOR each character in text:

IF the character is in the list of vowels:

INCREMENT vowel_count by 1.

PRINT vowel_count.

END

<!-- Add your pseudocode here -->

```python
text = input("Enter a string: ")

vowels = "aeiou"

vowel_count = 0

for char in text.lower():
    if char in vowels:
        vowel_count += 1

print(f"Number of vowels in the given string: {vowel_count}")
```

START

### Bonus Exercise 9: Find Factorial of a Number

Write pseudocode that calculates the factorial of a given positive integer.

STARTINPUT a positive integer and store it in n.

INITIALIZE a variable factorial to 

IF n is 0:

    SET factorial to 1 (by definition, $0! = 1$).
ELSE:

    FOR each number i from 1 up to and including n:

        MULTIPLY factorial by i.

PRINT the result.

END

```python
n = int(input("Enter a positive integer to find its factorial: "))

factorial = 1

if n < 0:
    print("Factorial does not exist for negative numbers.")
elif n == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, n + 1):
        factorial *= i
    print(f"The factorial of {n} is {factorial}")
```

### Bonus Exercise 10: FizzBuzz Problem

Write pseudocode for the FizzBuzz problem where you print numbers from 1 to N but replace multiples of three with "Fizz" and multiples of five with "Buzz".

START

INPUT a limit number and store it in N.

FOR each number i from 1 to N:

IF i is divisible by 3 AND i is divisible by 5:

PRINT "FizzBuzz"

ELSE IF i is divisible by 3:

PRINT "Fizz"

ELSE IF i is divisible by 5:

PRINT "Buzz"

ELSE:

PRINT the number i

END

<!-- Add your pseudocode here -->

```python
n = int(input("Enter the value of N: "))

for i in range(1, n + 1):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```
